```python
play_store_raw_df = pd.read_csv(r"C:\Users\yadelaja\Downloads\googleplaystore\googleplaystore.csv")
play_store_raw_df.loc[50:55]
```

Looking at the list of columns, I am uninterested in the size of the apps as well as the last 4 columns. I used the df.drop() method to get rid of those columns.

```python
column_names_raw = play_store_raw_df.columns

rename_columns = {}
for column in column_names_raw:
  rename_columns[column] = column.replace(' ','_').lower()

play_store_raw_df.rename(columns=rename_columns, inplace=True)
play_store_raw_df
```

```python
column_names = play_store_raw_df.columns
column_names
drop_columns = [column_names[i] for i in [4, 9, 10, 11, 12]]
play_store_raw_df.drop(drop_columns, inplace=True, axis=1)
play_store_raw_df
```

Now that we have loaded the data into a pandas dataframe, we can explore some introductory information about our dataset. From the columns in the dataframe, we can see that this dataset contains information about the app name, category, rating, number of reviews, install count etc. To get the size of the dataframe, we use the `df.shape` method. `df.info()` is also really useful to get a quick overview of the datatypes of the columns as well as the non-null counts which allows us to identify potentially incorrect/missing data fields. 

```python
play_store_raw_df.shape
```

```python
play_store_raw_df.info()
```

Using `df.shape` we can identify a couple of issues with the dataset:

*   Not all the columns have 10840 entries which indicates missing/incorrect data
*   columns like reviews, size, installs and price should be of the type 'int64/float64', however they are of type 'object'. This is a problem as we cannot perform numerical analysis on type 'object' data.

```python
play_store_raw_df['reviews'] = play_store_raw_df['reviews'].str.replace('.0M','0')

play_store_raw_df['installs'] = play_store_raw_df['installs'].str.replace('+','')
play_store_raw_df['installs'] = play_store_raw_df['installs'].str.replace(',','')
play_store_raw_df['installs'] = play_store_raw_df['installs'].str.replace('Free','0')

play_store_raw_df['price'] = play_store_raw_df['price'].str.replace('$','')
play_store_raw_df['price'] = play_store_raw_df['price'].str.replace('Everyone','')

play_store_raw_df[['reviews', 'installs', 'price']] = play_store_raw_df[['reviews', 'installs', 'price']].apply(pd.to_numeric,errors='ignore')
play_store_raw_df.info()
```

Another potential issue to check for is duplicate apps. We can do this by using the df.value_counts() method.

```python
play_store_raw_df.app.value_counts()
```

We can see that there are apps with duplicated rows in the dataset. For the purpose of this analysis, I want the latest updated version of the app. I can assume that the most updated version of the app data will be the one with the highest number of reviews.

```python
play_store_df_filtered = play_store_raw_df.iloc[play_store_raw_df.groupby("app")["reviews"].idxmax()]
play_store_df_filtered.app.value_counts()
```

```python
play_store_df = play_store_df_filtered.sort_index().reset_index(drop=True)
play_store_df['category'] = play_store_df.category.str.replace('_', ' ').str.title()
play_store_df
```
